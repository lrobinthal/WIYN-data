```python
%matplotlib notebook

import pandas
import numpy
import os
import matplotlib
import matplotlib.pyplot as plt
import datetime
from mpl_toolkits.mplot3d import Axes3D
import mpl_toolkits.axisartist.floating_axes as floating_axes
from matplotlib.projections import PolarAxes
```

```python
filename="WIYNqual1thru5.csv"
data=pandas.read_csv(filename)

data['ra']=data['RA']
data['dec']=data['Dec']
data['dist']=data['Distance']

data.info()
```

```python
tr = PolarAxes.PolarTransform()
ra0,ra1=numpy.radians(0), numpy.radians(360)
dis0,dis1=-3, 360
dis0, dis1=0, 800
#dis0, dis1=87,92
grid_helper = floating_axes.GridHelperCurveLinear(tr, extremes=(ra0, ra1, dis0, dis1))
print (grid_helper)
```

```python
fig=plt.figure()
fig.set_size_inches((6,6))

ax1=floating_axes.FloatingSubplot(fig, 111, grid_helper=grid_helper)  
fig.add_subplot(ax1)

ax1.axis["left"].set_axis_direction("bottom") #ra
ax1.axis["right"].set_axis_direction("top") #dist
#ax1.axis []

ax1.axis["left"].label.set_text(r"Distance")
ax1.axis["top"].label.set_text(r"RA")

aux_ax = ax1.get_aux_axes(tr)
aux_ax.patch = ax1.patch  
ax1.patch.zorder = 0.9  


aux_ax.scatter(data['ra'], data['dist'], data['dec'])
aux_ax.grid(True)
```

```python
r_cone=90.-data['dec']
angle_cone=90.-data['ra']
distance=data['dist']



f_x=numpy.tan(numpy.radians(numpy.sin(numpy.radians(angle_cone))*r_cone))*distance
f_y=numpy.tan(numpy.radians(numpy.cos(numpy.radians(angle_cone))*r_cone))*distance

#print(f_x)

fig=plt.figure()
ax=fig.add_subplot(111, projection='3d')
ax.scatter(f_x, f_y, distance, alpha=1, s=5)
ax.set_zlim((0, numpy.max(distance)))

ax.set_xlabel('Distance (Mpc)')
ax.set_ylabel('Distance (Mpc)')
ax.set_zlabel('Distance (Mpc)')

ax.set_ylim((-30,30))
ax.set_xlim((-30,30))
```
